```python
import numpy as np
import matplotlib.pyplot as plt

# Definim funcția f(x)
def f(x):
    return 4*x*(3 - x)

# Generăm valori pentru x între 0 și 6 pentru un grafic lin
x = np.linspace(0, 3, 100)
y = f(x)

# Punctele specificate de utilizator
x_points = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
y_points = f(x_points)

# Creăm plotul
plt.figure(figsize=(12, 8))
plt.plot(x, y, label=r'$v(t) = 4t(3-t)$', color='blue')  # Graficul funcției
plt.scatter(x_points, y_points, color='red', label='Puncte specifice', zorder=3)  # Punctele indicate

# Adăugăm etichete și titlu
plt.xlabel('t - timpul')
plt.ylabel('v(t) - viteza')
plt.title('v(t) = 4t(3-t)')
plt.legend()
plt.grid(True)

# Afișăm graficul
plt.show()
```

\( v'(t) = 12t - 4t^2 \)  
\( v'(t) = -4t^2 + 12t \)  
\( v'(t) = -8t + 12 \)  

\( v'(0.5) = 8 \)  
\( v'(1) = 4 \)  

\( v'(1.5) = 0 \)  

\( v'(2) = -4 \)  
\( v'(2.5) = -8 \)
